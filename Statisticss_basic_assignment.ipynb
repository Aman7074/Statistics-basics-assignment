Qn 1- What is statistics, and why is it important?

Ans- Statistics is the branch of mathematics that deals with collecting, organizing, analyzing, interpreting, and presenting data. It helps us make sense of large amounts of information by identifying patterns, relationships, and trends.

Qn 3- What are descriptive statistics?

Qn 4-  What is inferential statistics?

Instead of just describing data (like descriptive statistics), inferential statistics helps us make decisions or test hypotheses.



Qn 5- What is sampling in statistics?

Ans- What is Sampling in Statistics?
Sampling is the process of selecting a subset (sample) of individuals or observations from a larger population to analyze and make inferences about the entire population. Since studying an entire population is often impractical, sampling provides a more efficient way to gather data and draw conclusions.

 Nature: Quantitative (Numerical)


Qn 11- What is the measure of central tendency?

Ans- A measure of central tendency is a statistical term that describes a single value representing the center or typical value of a dataset. It helps summarize a large amount of data with a single value.

There are three main measures of central tendency:

* Mean (Average)
* Median (Middle Value)
* Mode (Most Frequent Value)

Qn 12- Define mean, median, and mode.

Ans-  

* Mean (Arithmetic Average)-

 Definition: The mean is the sum of all values divided by the total number of values.

 Example:

Data: 5, 10, 15, 20, 25

Mean: (5 + 10 + 15 + 20 + 25) ÷ 5 = 15


* Median (Middle Value)-

 Definition: The median is the middle value when a dataset is arranged in ascending order.

 Steps to Find the Median:

Sort the data in ascending order.

If odd number of values → Middle value is the median.

If even number of values → Median is the average of two middle values.

* Mode (Most Frequent Value)-

 Definition: The mode is the most frequently occurring value in a dataset.

 Example:

Data: 3, 5, 7, 7, 9, 11

Mode: 7 (appears twice, while others appear once)

Qn 14- What is variance, and how is it calculated?

Ans- Variance measures how spread out the values in a dataset are relative to the mean. It tells us how much individual data points differ from the average.

*  High variance → Data points are widely spread out (more variation).

* Low variance → Data points are closely clustered around the mean (less variation).



Qn 15- What is standard deviation, and why is it important?

Ans- Standard deviation measures how spread out the data points are relative to the mean. It is the square root of variance and is used to express variation in the same unit as the data.

 High standard deviation → Data points are widely spread out.

 Low standard deviation → Data points are close to the mean.

Qn 19-  What does it mean if a dataset is positively or negatively skewed?

Ans-

* Positively Skewed (Right-Skewed) Distribution-

Tail extends to the right (toward higher values).

Mean > Median > Mode (the mean is pulled right by extreme high values).

Most data points are concentrated on the left.

* Negatively Skewed (Left-Skewed) Distribution


Tail extends to the left (toward lower values).

Mean < Median < Mode (the mean is pulled left by extreme low values).

Most data points are concentrated on the right.

Qn 20- Define and explain kurtosis?

Ans- Kurtosis measures the "tailedness" of a data distribution. It describes how peaked or flat the distribution is compared to a normal distribution.
📌 Key Concept:

High kurtosis -- More extreme outliers (fatter tails).

Low kurtosis -- Fewer extreme outliers (thinner tails).

 Types of Kurtosis

1- Mesokurtic (Normal Kurtosis)

A normal distribution has a kurtosis of 3.

Example: Standard Normal Distribution.


2- Leptokurtic (High Kurtosis)

Sharper peak and fatter tails than a normal distribution.

More extreme values (outliers).

Example: Stock market crashes, financial returns.


3- Platykurtic (Low Kurtosis)

Flatter peak and thinner tails than a normal distribution.

Fewer extreme values.

Example: Uniform distribution.




Qn 21- What is the purpose of covariance?

Qn 22- What does correlation measure in statistics?

Qn 23- What is the difference between covariance and correlation?

Qn 1- How do you calculate the mean, median, and mode of a dataset?

```python
# Answer 1

import statistics

data = [10, 20, 30, 40, 50]

mean_value = statistics.mean(data)
median_value = statistics.median(data)
mode_value = statistics.mode([2, 3, 3, 5, 7, 7, 7, 9])

print(f"Mean: {mean_value}")
print(f"Median: {median_value}")
print(f"Mode: {mode_value}")
```

Qn 2- Write a Python program to compute the variance and standard deviation of a dataset.

```python
# Answer 2

data = [10, 20, 30, 40, 50]

# Compute variance and standard deviation
variance_value = statistics.variance(data)  # Sample variance (n-1)
std_dev_value = statistics.stdev(data)  # Sample standard deviation (n-1)

# Print results
print(f"Variance: {variance_value}")
print(f"Standard Deviation: {std_dev_value}")
```

Qn 4- Implement sampling techniques like random sampling and stratified sampling.

```python
# Answer 4

import random
import pandas as pd
from sklearn.model_selection import train_test_split

# Sample dataset
data = {
    "ID": range(1, 21),  # 20 unique IDs
    "Category": ["A", "B", "A", "C", "B", "C", "A", "A", "B", "C",
                 "A", "B", "A", "C", "B", "C", "A", "B", "C", "A"],
    "Value": [random.randint(10, 100) for _ in range(20)]  # Random values
}

df = pd.DataFrame(data)


random_sample = df.sample(n=5, random_state=42)


stratified_sample, _ = train_test_split(df, test_size=0.75, stratify=df["Category"], random_state=42)


print("🔹 Random Sample:\n", random_sample)
print("\n🔹 Stratified Sample:\n", stratified_sample)
```

Qn 5- 5 Write a Python function to calculate the range of a dataset.

```python
# Answer 5

def calculate_range(data):
    """Returns the range of a dataset (max - min)."""
    if not data:
        return "Dataset is empty!"

    return max(data) - min(data)


dataset = [15, 22, 8, 36, 42, 10]

# Calculate and print the range
range_value = calculate_range(dataset)
print(f"Range of dataset: {range_value}")
```

Qn 8- Generate a dataset and demonstrate positive and negative skewness.

```python
# Answer 8


import seaborn as sns
import scipy.stats as stats

positively_skewed = np.random.exponential(scale=2, size=1000)  # Right-skewed (Positive)
negatively_skewed = np.random.normal(loc=50, scale=10, size=1000) ** 2  # Left-skewed (Negative)

pos_skewness = stats.skew(positively_skewed)
neg_skewness = stats.skew(negatively_skewed)

# Plot histograms
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Positive Skew
sns.histplot(positively_skewed, bins=30, kde=True, color='blue', ax=axes[0])
axes[0].set_title(f"Positive Skew (Skewness = {pos_skewness:.2f})")

# Negative Skew
sns.histplot(negatively_skewed, bins=30, kde=True, color='red', ax=axes[1])
axes[1].set_title(f"Negative Skew (Skewness = {neg_skewness:.2f})")

plt.tight_layout()
plt.show()
```

Qn 9- Write a Python script to calculate covariance between two datasets.

```python
# Answer 9

x = [10, 12, 14, 16, 18, 20]
y = [55, 60, 65, 70, 75, 80]


cov_matrix = np.cov(x, y, bias=False)
covariance = cov_matrix[0, 1]


df = pd.DataFrame({'Study_Hours': x, 'Exam_Scores': y})
cov_pandas = df.cov().iloc[0, 1]


print(f"Covariance (NumPy): {covariance:.2f}")
print(f"Covariance (Pandas): {cov_pandas:.2f}")
```

Qn 12- Implement and compare simple random sampling and systematic sampling.

```python
# Answer 12

np.random.seed(42)
population_size = 100
data = pd.DataFrame({'Student_ID': range(1, population_size + 1),
                     'Test_Score': np.random.randint(50, 100, population_size)})


sample_size = 10


srs_sample = data.sample(n=sample_size, random_state=42)


interval = population_size // sample_size
sys_sample = data.iloc[::interval]


print(" Simple Random Sampling:\n", srs_sample)
print("\ Systematic Sampling:\n", sys_sample)
```

Qn 23- Visualize skewness and kurtosis using Python libraries like matplotlib or seaborn.


```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats


np.random.seed(42)
data = np.random.gamma(shape=2, scale=2, size=1000)

skewness = stats.skew(data)
kurtosis = stats.kurtosis(data)


plt.figure(figsize=(8,5))
sns.histplot(data, kde=True, bins=30, color="blue", edgecolor="black")


plt.axvline(np.mean(data), color='red', linestyle='dashed', label="Mean")
plt.text(np.mean(data) + 0.5, 50, f"Skewness: {skewness:.2f}", color="red")
plt.text(np.mean(data) + 0.5, 40, f"Kurtosis: {kurtosis:.2f}", color="blue")


plt.title("Histogram Showing Skewness & Kurtosis", fontsize=14)
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

Qn 24- Implement the Pearson and Spearman correlation coefficients for a dataset.

```python
# Answer 24


np.random.seed(42)
x = np.random.randint(50, 100, 100)
y = x + np.random.normal(0, 10, 100)


df = pd.DataFrame({"Feature X": x, "Feature Y": y})


pearson_corr, _ = stats.pearsonr(df["Feature X"], df["Feature Y"])
spearman_corr, _ = stats.spearmanr(df["Feature X"], df["Feature Y"])


print(f" Pearson Correlation: {pearson_corr:.2f}")
print(f" Spearman Correlation: {spearman_corr:.2f}")


plt.figure(figsize=(7,5))
sns.scatterplot(x=df["Feature X"], y=df["Feature Y"], color="blue")
plt.title("Scatter Plot of Feature X vs Feature Y", fontsize=14)
plt.xlabel("Feature X")
plt.ylabel("Feature Y")


plt.text(55, max(y)-5, f"Pearson: {pearson_corr:.2f}", color="red", fontsize=12)
plt.text(55, max(y)-10, f"Spearman: {spearman_corr:.2f}", color="blue", fontsize=12)

plt.show()
```
